### Identifying Numerical and categorical columns:

```python
#Numerical data:

numeric_cols = inputs_df.select_dtypes(include= ['int64', 'float64']).columns.tolist()
```

```python
categorical_cols = inputs_df.select_dtypes(include='object').columns.tolist()
```

```python
print(list(numeric_cols))
```

```python
print(categorical_cols)
```

#### Imputing missing data in Numerical columns:

```python
missing_vals = inputs_df[numeric_cols].isna().sum().sort_values(ascending=False)
```

```python
print(missing_vals[missing_vals>0])
```

we have to impute values in these 3 cols:


we will impute our missing datas using the SimpleImputer from sklearn.impute

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy= 'mean')
```

```python

imputer.fit(inputs_df[numeric_cols])
```

```python
inputs_df[numeric_cols] = imputer.transform(inputs_df[numeric_cols])
```

```python
inputs_df[numeric_cols].isna().sum().sort_values(ascending = False)
```

all our missing values are filled in now


Now we can proceed with scaling down the features

### Scaling down the features

```python
inputs_df[numeric_cols]
```

```python
inputs_df[numeric_cols].describe().loc[['min', 'max']]
```

our data has very large variating quantities, we have to scale down these features or it will end up impacting our model.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(inputs_df[numeric_cols])
```

```python
inputs_df[numeric_cols] = scaler.transform(inputs_df[numeric_cols])
```

```python
inputs_df[numeric_cols]
```

now we have scaled our features down to (0,1) range:

### Encoding our categorical columns using OneHotEncoder:


```python
inputs_df[categorical_cols]
```

```python
inputs_df[categorical_cols].nunique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
```

```python
encoder.fit(inputs_df[categorical_cols])
```

```python
encoded_cols = list(encoder.get_feature_names(categorical_cols))
print(encoded_cols)
```

```python
inputs_df[encoded_cols] = encoder.transform(inputs_df[categorical_cols])
```

```python
inputs_df[encoded_cols]
```

```python
inputs_df
```

### Splitting our dataset into Training, Validation Datasets:

```python
from sklearn.model_selection import train_test_split
```

```python


train_input, val_input, train_target, val_target = train_test_split(inputs_df[numeric_cols + encoded_cols], targets, test_size = 0.25, random_state = 42)
```

### Creating a LinearRegression model using Ridge

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge()
```

```python
model.fit(train_input, train_target)
```

Evaluating the created model, by predicting a new set of values and computing the Mean RMS Error

```python
from sklearn.metrics import mean_squared_error
```

```python
train_preds = model.predict(train_input)
```

```python
train_preds
```

```python
train_rmse = mean_squared_error(train_target, train_preds, squared=False)
```

```python
print("The RMSE loss on the Training Set: ${}".format(train_rmse))
```

checking on validation set:

```python
val_preds = model.predict(val_input)
```

```python
val_preds
```

```python
val_rmse = mean_squared_error(val_target, val_preds, squared= False)
```

```python
print("RMSE on validation set: ${}".format(val_rmse))
```

```python
#Creating a helper function to predict new custom inputs




def predict_new(single_input):

    #Convert the input into a dataframe
    input_df = pd.DataFrame([single_input])

    #Filling in all the missing values
    input_df[numeric_cols] = imputer.transform(input_df[numeric_cols])

    #Scaling them down
    input_df[numeric_cols] = scaler.transform(input_df[numeric_cols])

    #Encoding the categorical columns
    input_df[encoded_cols] = encoder.transform(input_df[categorical_cols].values)

    #Appending the new changed columns into a new dataset
    X_input = input_df[numeric_cols+encoded_cols]

    return model.predict(X_input)[0]
```
